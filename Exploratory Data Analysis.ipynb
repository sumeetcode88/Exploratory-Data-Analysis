# <u>Exploratory Data Analysis</u>

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

<b>Using this dataset, we can build a classification system which can predict whether a person has Benign or Malignant tumor. Malignant tumors are considered cancerous. In the EDA part, we will try to understand the characteristics of the data and its descriptive measures.</b>

```python
breast_cancer_data=pd.read_csv(r"E:\Data science projects\EDA (Exploratory Data Analysis)\breast_cancer_data.csv")
#import dataset
```

<b>Data copied</b> so that original data does get manipulate and in future if needed we can refer original data for study

```python
data=breast_cancer_data.copy()
```

### 1. Checking all features in the dataset

```python
data.head()
```

Data has 569 rows (data points) and 32 columns (features)

```python
data.shape
```

### 2. Checking data type of each column and non-null count

```python
data.info()
```

We found columns "id" as integer, "diagnosis" as object i.e. categorical and rest all columns as continuous numerical variables

The second column in data is “diagnosis”, where, “M” represents Malignant & “B” represents Benign. <br><b>“diagnosis”</b> is our <b>Target column</b>.

This is a <b>Classification</b> probem. We need to encode "diagnosis" column into 1 and 0 so that all columns are in numerical format.<br>
<b>Label encoding uses alphabetical ordering. Hence, "B" will be encoded with 0 and "M' with 1.</b>

### 3. Encoding the labels for classification problems

```python
label_encode = LabelEncoder()
labels = label_encode.fit_transform(data['diagnosis'])
data['target'] = labels
data.drop(columns=['id','diagnosis'], axis=1, inplace=True)
data.head()
```

Above we encoded the column 'diagnosis' and stored in a new column 'target'. I have removed the columns 'id' and 'diagnosis' as we do not need them.

### 4. Checking for missing values

```python
data.isnull().sum()
```

Above code line finds null values and sum the values for each column.<br> In our dataset there are no missing values.

### 5. Descriptive Statistics of the dataset

```python
data.describe()
```

describe() function in pandas is used to get some statistical measures about the dataset.<br> This is what we call as <b>“Descriptive Statistics”</b> which is a summarization of the data.<br>The main inference that we can get here is, for most of the columns, the <b>mean value</b> is larger than <b>median value</b> (50th percentile: 50%). <br>This is an indication that those features have a right skewed data. This information will be visible for us when we create distribution plot for individual features in Data Visualization part.

### 6. Checking the distribution of the target variable:

```python
data['target'].value_counts()
```

Here we are checking the distribution of the dataset based on the target variable to see if there is an imbalance. This is an exclusive step for <b>Classification problems</b>.<br>
The output shows slight imbalance in the dataset number of <b>Benign(0)</b> cases are more than number of <b>Malignant(1)</b> cases), but not too much to worry about in this case.

### 7. Grouping the data based on target variable

This step is also exclusive for Classification problems.<br>
This is to group the dataset based on the target variable. <br>
We will be grouping the data points as 0 & 1 representing Benign & Malignant respectively.<br>
This grouping is done with the mean value of all the columns.

We see that the mean value for most of the features are greater for Malignant cases than the mean value for Benign cases. <b>This inference is very important.</b>

## Data Visualization is a part of Exploratory Data Analysis(EDA). 

Importing the Data Visualization Libraries.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Matplotlib & Seaborn are the two main Data Visualization libraries in Python. There are also other libraries like Plotly and GGplot.

### 1. Count plot for Categorical columns:

This dataset contains only one categorical variable(“target”) with two categories: 0 (Benign) and 1(Malignant). When we have <b>categorical variable</b>, we will plot it in a <b>count plot</b> and when we have a <b>numerical variable</b>, we will use a <b>distribution plot</b>.

```python
sns.countplot(x='target', data=data)
```

We can see that the number of data points with label “0” is higher than label “1”.<br> This means that we have more <b>Benign cases</b> compared to <b>Malignant cases</b> in the dataset. So we can say that this dataset is slightly imbalanced.<br>
<i>Count plot will show the total counts in each category.</i>

### 2. Distribution plot for all columns

Since all columns contain numerical values we can build distribution plot.<br>
<i>Distribution plot tells us whether the data is Normally Distributed or there is some Skewness in the data.</i>

When the skewness in the data is large, we may need to do some transformations, in order to get better results from the Machine Learning models once we train them.

```python
plt.figure(figsize=(15,40))
for i, column in enumerate(data.columns, 1):
    plt.subplot(10,4,i)
    sns.distplot(data[column], bins=15)
    #plt.xticks(rotation=45)
```

```python
#This will show countplot which has similar observations as above
plt.figure(figsize=(15,30))
for i, column in enumerate(data.columns, 1):
    plt.subplot(10,4,i)
    sns.countplot(column, data=data)
    #plt.xticks(rotation=45)
```

Most of the columns have right skewness in this dataset. <br>
We will deal with this in the Feature Engineering part.

This is an what we call as Univariate Analysis where we are take one variable at a time and analyzing it. When we take two variables at a time and try to find the relationship between them, it is called Bivariate Analysis. When we have more than two variables, it is called Multivariate Analysis.

### 4. Checking for Outliers

Outliers detection is one of the important tasks that we have to do. Most of the Machine Learning models like Regression models, K-Nearest Neighbors, etc. are sensitive to outliers. On the other hand, models like Random Forest are not affected by Outliers. So we should process these outliers before feeding the data to those models. This will also be carried out in the Feature Engineering part.

```python
for column in data:
   plt.figure()
   data.boxplot([column])
```

The circles above the top whisker and below the bottom whisker represents the Outliers.<br>
Here we have <b>outliers</b> in the <b>upper range alone</b>. 

### 5. Correlation matrix

The main purpose of a correlation matrix is to understand the correlation (in other words, relationship) between the variables present in a dataset. It is very helpful in Feature Selection which is carried out to choose the important features and remove the irrelevant and unnecessary features.

We use "data.corr()" to pull Correlation matrix which finds relationship between variables.<br>
We are creating a Heat Map to visualize the correlation between the variables.

```python
correlation_matrix = data.corr()
plt.figure(figsize=(20,20))
sns.heatmap(correlation_matrix, cbar=True, fmt='.1f', annot=True, cmap='Blues')
plt.savefig('Correlation Heat map')
```

<i>The correlation values ranges from +1 to -1. If the correlation between two variables is +1, it means that the value of one variable increases by 1 unit, if the other variable increases by 1 unit and vice versa. This is called <b>Positive Correlation</b>.<br>
If the correlation between two variables is -1, it means that the value of one variable decreases by 1 unit if the other variable increases by 1 unit and vice versa. This is called <b>Negative Correlation</b>.</i><br>
<b>In simple words Positive Correlation is direct proportionality & Negative Correlation is inverse proportionality.</b> 

When we have two independent variables that are very highly correlated, we should remove one of them because we run into the <b>multicollinearity problem</b>. In those cases, coefficients related to the two highly correlated variables will be unreliable.
